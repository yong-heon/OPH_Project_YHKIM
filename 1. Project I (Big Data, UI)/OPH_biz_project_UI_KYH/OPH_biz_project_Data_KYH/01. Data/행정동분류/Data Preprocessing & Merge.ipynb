```python
## Lib import
import pandas as pd
from google.colab import files

## 파일선택으로 업로드
uploaded_files = files.upload()

df_monthlyAvgAPT = pd.read_excel('아파트월세&보증금.xlsx')
df_monthlyAvgMEN = pd.read_excel('다세대월세&보증금.xlsx')
df_monthlyAvgSTU = pd.read_excel('오피스텔월세&보증금.xlsx')

df_transferPopulation = pd.read_excel('전입인구.xlsx')
df_totalPopulation = pd.read_excel('인구밀도.xlsx')

df_saleAvgAPT = pd.read_excel('아파트매매.xlsx')
df_saleAvgMEN = pd.read_excel('다세대매매.xlsx')
df_saleAvgSTU = pd.read_excel('오피스텔매매.xlsx')

df_shoppingmall = pd.read_excel('대형마트.xlsx')
df_emergencyRM = pd.read_excel('응급실.xlsx')
df_park = pd.read_excel('공원.xlsx')
df_subway = pd.read_excel('지하철역.xlsx')

df_jeonseAvgAPT = pd.read_excel('아파트전세.xlsx')
df_jeonseAvgMEN = pd.read_excel('다세대전세.xlsx')
df_jeonseAvgSTU = pd.read_excel('오피스텔전세.xlsx')


## 엑셀파일을 SGG_NM과 BJDONG_NM이 같은 column들끼리 종합
merged_df = pd.merge(df_monthlyAvgAPT, df_monthlyAvgMEN, on=['SGG_NM', 'BJDONG_NM'], how='outer')
merged_df = pd.merge(merged_df, df_monthlyAvgSTU, on=['SGG_NM', 'BJDONG_NM'], how='outer')

merged_df = pd.merge(merged_df, df_transferPopulation,  on=['SGG_NM', 'BJDONG_NM'], how='outer')
merged_df = pd.merge(merged_df, df_totalPopulation, on=['SGG_NM', 'BJDONG_NM'], how='outer')
merged_df = pd.merge(merged_df, df_saleAvgAPT,  on=['SGG_NM', 'BJDONG_NM'], how='outer')
merged_df = pd.merge(merged_df, df_saleAvgSTU, on=['SGG_NM', 'BJDONG_NM'], how='outer')
merged_df = pd.merge(merged_df, df_saleAvgMEN, on=['SGG_NM', 'BJDONG_NM'], how='outer')

merged_df = pd.merge(merged_df, df_emergencyRM,  on=['SGG_NM', 'BJDONG_NM'], how='outer')
merged_df = pd.merge(merged_df, df_shoppingmall,  on=['SGG_NM', 'BJDONG_NM'], how='outer')
merged_df = pd.merge(merged_df, df_park,  on=['SGG_NM', 'BJDONG_NM'], how='outer')
merged_df = pd.merge(merged_df, df_subway, on=['SGG_NM', 'BJDONG_NM'], how='outer')

merged_df = pd.merge(merged_df, df_jeonseAvgAPT, on=['SGG_NM', 'BJDONG_NM'], how='outer')
merged_df = pd.merge(merged_df, df_jeonseAvgMEN, on=['SGG_NM', 'BJDONG_NM'], how='outer')
merged_df = pd.merge(merged_df, df_jeonseAvgSTU, on=['SGG_NM','BJDONG_NM'], how='outer')


## merged된 데이터프레임 csv로 전환
merged_df.to_csv('서울시행정동별통합데이터.csv', index=False)
```

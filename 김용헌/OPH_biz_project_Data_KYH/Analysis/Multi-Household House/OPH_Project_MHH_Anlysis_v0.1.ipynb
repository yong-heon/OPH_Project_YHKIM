```python
# 지역데이터 불러오기
sgg = pd.read_csv("/content/drive/MyDrive/OPH_Project_Data/서울시행정동별통합데이터.csv")


# 자치구 열을 기준으로 그룹화하여 각 열의 합계 및 평균 구하기
sgg_grouped = sgg.groupby('자치구').agg({
    '인구전입량': 'sum',
    '청년1인가구인구수': 'sum',
    '응급실수': 'sum',
    '대형마트수': 'sum',
    '주요공원수': 'sum',
    '지하철역수': 'sum',
    '아파트월세보증금': 'mean',
    '아파트월세': 'mean',
    '다세대주택월세보증금': 'mean',
    '다세대주택월세': 'mean',
    '오피스텔월세보증금': 'mean',
    '오피스텔월세': 'mean',
    '아파트매매가': 'mean',
    '오피스텔매매가': 'mean',
    '다세대주택매매가': 'mean',
    '아파트전세': 'mean',
    '다세대주택전세': 'mean',
    '오피스텔전세': 'mean'
}).reset_index()

# 결과 출력
display(sgg_grouped)

sgg_grouped.to_csv('서울시자치구별통합데이터.csv', index=False)
```

## 자치구별 다세대주택 부동산 가격 표준편차 & 평균 계산
---

```python
import pandas as pd
from scipy.stats import zscore

# 다세대주택 가격 데이터만 추출하여 새로운 데이터프레임 생성
df_MHH_price = dong_grouped.copy()
df_std_devs = df_MHH_price[['다세대주택매매가', '다세대주택전세', '다세대주택월세보증금', '다세대주택월세']].copy()


 # 전체 자치구에 대한 다세대주택 매매가, 다세대주택 전세가, 다세대주택 월세의 표준편차를 계산
std_devs_trade = np.std(df_MHH_price['다세대주택매매가'])
std_devs_lease = np.std(df_MHH_price['다세대주택전세'])
std_devs_rentdeposit = np.std(df_MHH_price['다세대주택월세보증금'])
std_devs_rent = np.std(df_MHH_price['다세대주택월세'])


# std_devs 들을 pd로 변환해야 됨
std_devs_trade = pd.Series(np.std(df_MHH_price['다세대주택매매가']), name='다세대주택매매가')
std_devs_lease = pd.Series(np.std(df_MHH_price['다세대주택전세']), name='다세대주택전세')
std_devs_rentdeposit = pd.Series(np.std(df_MHH_price['다세대주택월세보증금']), name='다세대주택월세보증금')
std_devs_rent = pd.Series(np.std(df_MHH_price['다세대주택월세']), name='다세대주택월세')


# 결과를 하나의 DataFrame으로 통합
std_result_MHH = pd.concat([std_devs_trade, std_devs_lease, std_devs_rentdeposit, std_devs_rent], axis=1)

# 결과 출력
display(std_result_MHH)

# 다세대주택 가격 데이터 평균 계산
df_MHH_selected = dong_grouped.copy()[['다세대주택매매가', '다세대주택전세', '다세대주택월세보증금', '다세대주택월세']]
MHH_means = df_MHH_selected.mean()

# 계산한 평균을 데이터프레임으로 변환
df_MHH_means = pd.DataFrame(MHH_means).transpose()

# 결과확인
display(df_MHH_means)
```

## 자치구별 아파트 부동산 가격 z-score로 계산
---

```python
# 자치구별 다세대주택 가격 데이터 표준점수 (Z-score) 계산
df_MHH_z_scores_no_sgg = (dong_grouped[['다세대주택매매가', '다세대주택전세', '다세대주택월세보증금', '다세대주택월세']].copy() - df_MHH_means.values) / std_result_MHH.values

# 결과 출력
(df_MHH_z_scores_no_sgg)

# '자치구' 컬럼 선택
df_sgg = dong_grouped[['자치구']]

# '자치구' 컬럼과 표준점수 데이터프레임 연결
df_MHH_z_scores = pd.concat([df_sgg, df_MHH_z_scores_no_sgg], axis=1)

# 결과 출력
display(df_MHH_z_scores)
```

## 표준점수를 백분위 점수로 변환
---

```python
# import lib
from scipy.stats import norm

# Z-Score를 Percentile Rank로 변환
df_percentiles_MHH = df_MHH_z_scores_no_sgg.apply(lambda x: norm.cdf(x) * 100)

# 소수점 한 자리까지 반올림
df_percentiles_MHH = df_percentiles_MHH.round(1)

# '%' 기호 추가
df_percentiles_MHH = df_percentiles_MHH.astype(str) + '%'

# '자치구' 컬럼과 표준점수 데이터프레임 연결
df_MHH_z_scores = pd.concat([df_sgg, df_percentiles_MHH], axis=1)

# 결과 출력
display(df_MHH_z_scores)

# csv 파일 변환
df_MHH_z_scores.to_csv('서울시_자치구별_다세대주택가격_종합_백분위.csv', index=False)
```
